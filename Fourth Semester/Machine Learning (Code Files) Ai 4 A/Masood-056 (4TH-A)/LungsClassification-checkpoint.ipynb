Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

from imblearn.over_sampling import SMOTE
from scipy.stats import zscore
```

Loading Dataset

Cleaning Column Names

```python
df.columns = df.columns.str.replace(" ", "_")
```

Outlier Removal

```python
z_scores = np.abs(zscore(df.select_dtypes(include=[np.number])))
df = df[(z_scores < 3).all(axis=1)]
```

Separation

```python
X = df.drop("LUNG_CANCER", axis=1)
y = df["LUNG_CANCER"].map({"YES": 1, "NO": 0})
```

 Preprocessing Setup

```python
categorical_features = ["GENDER"]
numerical_features = X.drop(columns=categorical_features).columns.tolist()
```

Preprocessing 

```python
preprocessor = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first'), categorical_features),
    ('scaler', StandardScaler(), numerical_features)
])
```

Apply Preprocessing

```python
X_processed = preprocessor.fit_transform(X)
```

Balancing the Dataset with SMOTE

```python
X_resampled, y_resampled = SMOTE(random_state=42).fit_resample(X_processed, y)
```

Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

Model Training

```python
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
```

Model Evaluation

Predict Probabilities

```python
y_pred = clf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
y_proba = clf.predict_proba(X_test)[:, 1]
```

Prediction Graph

```python
plt.figure(figsize=(8, 6))
sns.histplot(y_proba, bins=20, kde=True, color='darkorange')
plt.title("Distribution of Predicted Probabilities for Lung Cancer (Class = 1)")
plt.xlabel("Predicted Probability")
plt.ylabel("Number of Instances")
plt.tight_layout()
plt.show()
```
