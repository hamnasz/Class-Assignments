##Clean Data

```python
data.isnull().sum()
```

```python
data.drop_duplicates(inplace=True)
data.shape
```

##Outlier Removal

```python
numerical_cols = ['Square Footage', 'Number of Occupants', 'Appliances Used', 'Average Temperature']
```

```python
Q1 = data[numerical_cols].quantile(0.25)
Q3 = data[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

outlier_removed_data = data[~((data[numerical_cols] < (Q1 - 1.5 * IQR)) | (data[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

outlier_removed_data.head()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
data.boxplot()
plt.title("Before Outlier Removal")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 2)
outlier_removed_data.boxplot()
plt.title("After Outlier Removal")

plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['Building Type', 'Day of Week']

encoded_data = pd.get_dummies(outlier_removed_data, columns=categorical_cols)

encoded_data.head()
```

```python
x = encoded_data.drop('Energy Consumption', axis=1)

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

feature_names = x.columns
x_scaled = pd.DataFrame(x_scaled, columns=feature_names)
```

```python
plt.figure(figsize=(20, 6))
sns.boxplot(data=x_scaled)
plt.title('Scaled Numerical Features')
```

```python
x = x_scaled
y = encoded_data['Energy Consumption']
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(y, bins=30, kde=True)
plt.title('Target Variable Distribution After Outlier Removal')
plt.xlabel('Energy Consumption (kWh)')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("\nEvaluation Metrics:")
print(f"MAE:  {mae:.4f}")
print(f"MSE:  {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
```

```python
results = pd.DataFrame({"Actual": y_test.values, "Predicted": y_pred})
print("\nSample Predictions:\n", results.head())
```
